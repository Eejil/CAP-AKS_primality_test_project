```python
def perfect_pow(n):
    # We want to find integers a, b with a**b = n
    for b in range(2, floor(log(n,2))):
        # Given b and n, we have a = n**(1/b)
        a = n**(1/b)
        # a is a float, so we need to check this is a valid integer
        if int(a) == a:
            return 1
    return n     

def find_r(n):
    maxk = floor(log(n,2)^2);
    maxr = max(3,ceil(log(n,2)^5));
    nextR = True;
    k = 1;
    r = 2;
    while nextR == True and r < maxr:
        nextR = False;
        r = r + 1;
        while nextR == False and k <= maxk:
            nextR = (mod(n^k,r) == 1 or mod(n^k,r) == 0)
            k = k+1;
    r = r-1;
    return r; 

def aks_fun(n):
    if perfect_pow(n) == 1:
        return "composite" 
    r = find_r(n);

    for a in range(2,min(r+1,n)):
        if mod(n,a) == 0:
            return "composite"
    if n <= r: 
        return "prime"
    Z = Integers(n) 
    R = PolynomialRing(Z, 'X')
    X = R.0
    I = X^r-1;
    I = R.ideal(I)
    R = sage.rings.quotient_ring.QuotientRing(R, I,'X')
    for b in range(1,floor(math.sqrt(euler_phi(r)*log(n,2)))):
        if R((X+b)^n) != X^n+b:
            return "composite"
    return "prime"
```
